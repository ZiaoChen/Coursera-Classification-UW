#### Read Data

```python
import sframe
import numpy as np
import json
products = sframe.SFrame('data2/')
with open ('data2/important_words.json') as data:
    important_words = json.load(data)
    important_words = [str(word) for word in important_words]
```

#### Preprocessing

```python
def remove_punctuation(text):
    import string
    return text.translate(None,string.punctuation)

products['review_clean'] = products['review'].apply(remove_punctuation)

for word in important_words:
    products[word] = products['review_clean'].apply(lambda s: s.split().count(word))
products
```

#### Split training and validation data

```python
train_data, validation_data = products.random_split(.8,seed=2)
print 'Training set   : %d data points' % len(train_data)
print 'Validation set : %d data points' % len(validation_data)
```

#### Convert SFrame to Numpy array

```python
def get_numpy_data(data_sframe,features,label):
    data_sframe["intercept"] = 1
    features = ["intercept"] + features
    features_sframe = data_sframe[features]
    feature_matrix = features_sframe.to_numpy()
    label_sarray = data_sframe[label]
    label_array = label_sarray.to_numpy()
    return (feature_matrix,label_array)

feature_matrix_train, sentiment_train = get_numpy_data(train_data, important_words,'sentiment')
feature_matrix_valid, sentiment_valid = get_numpy_data(validation_data, important_words,'sentiment')
```

#### Predict probability function

```python
def predict_probability(feature_matrix, coefficients):
    predictions = 1 / (1 + np.exp(-np.dot(feature_matrix,coefficients)))
    return predictions
```

#### Adding L2 penalty

```python
def feature_derivative_with_L2(errors, feature, coefficient, l2_penalty, feature_is_constant):
    derivative = np.dot(feature,errors)
    
    if not feature_is_constant:
        derivative = derivative - 2*l2_penalty*coefficient
    return derivative
```

#### Compute log likelihood function

```python
def compute_log_likelihood_with_L2(feature_matrix,sentiment,coefficients,l2_penalty):
    indicator = (sentiment==+1)
    scores = np.dot(feature_matrix, coefficients)
    lp = np.sum((indicator-1)*scores-np.log(1. + np.exp(-scores))) - l2_penalty*np.sum(coefficients[1:]**2)
    
    return lp
```

#### Model function

```python
def logistic_regression_with_L2(feature_matrix, sentiment, initial_coefficients,step_size,l2_penalty,max_iter):
    coefficients = np.array(initial_coefficients)
    for itr in xrange(max_iter):
        predictions = predict_probability(feature_matrix,coefficients)
        
        indicator = (sentiment == +1)
        
        errors = indicator - predictions
        
        for j in xrange(len(coefficients)):
            is_intercept = (j==0)
            derivative = feature_derivative_with_L2(errors,feature_matrix[:,j],coefficients[j],l2_penalty,is_intercept)
            coefficients[j] = coefficients[j] + step_size*derivative
        if itr <= 15 or (itr <= 100 and itr % 10 == 0) or (itr <= 1000 and itr % 100 == 0) or (itr <= 10000 and itr % 1000 == 0) or itr % 10000 == 0:
            lp = compute_log_likelihood_with_L2(feature_matrix, sentiment, coefficients, l2_penalty)
            print 'iteration %*d: log likelihood of observed labels = %.8f' % (int(np.ceil(np.log10(max_iter))), itr, lp) 
    return coefficients
```

#### L2 = 0

```python
coefficients_0_penalty = logistic_regression_with_L2(feature_matrix_train, sentiment_train,initial_coefficients = np.zeros(194),step_size = 5e-6, l2_penalty=0,max_iter=501)
```
