#### Read data, subset of the Amazon Product Review Dataset

```python
import numpy as np
```

```python
import sframe
products = sframe.SFrame('data2/').to_dataframe()
products
```

#### Simple data transformation 

Fill na values in 'review' column 

```python
products = products.fillna({'review':""})
```

Remove Punctuation 

```python
def remove_punctuation(text):
    import string
    return text.translate(None,string.punctuation)

products['review_clean'] = products['review'].apply(remove_punctuation)
```

#### Read important words 

```python
import json
with open ('data2/important_words.json') as data:
    important_words = json.load(data)
important_words = [str(word) for word in important_words]
important_words
```

#### Count number of appearance of important words for each review 

```python
for word in important_words:
    products[word] = products['review_clean'].apply(lambda s: s.split().count(word))
products
```

#### Convert dataframe to multi-dimensional array

```python
 def get_numpy_data(dataframe, features, label):
     
    #Add constant column
    #Need to make sure 'constant' is the first column
    dataframe['constant'] = 1
    features = ['constant'] + features
    
    #Convert data into 2d array
    features_frame = dataframe[features]
    feature_matrix = features_frame.as_matrix()
    
    #Convert label into 1d array
    label_sarray = dataframe[label]
    label_array = label_sarray.as_matrix()
    
    return (feature_matrix, label_array)

feature_matrix, sentiment = get_numpy_data(products,important_words,'sentiment')
```

#### Estimate conditional Probability

```python
def predict_probability(feature_matrix, coefficient):
    score = np.dot(feature_matrix,coefficient)
    predictions = 1 / (1 + np.exp(-score))
    return predictions
```

#### Compute derivative (dot product of error and feature value)

```python
def feature_derivative(errors, feature):
    derivative = np.dot(errors,feature)    
    return derivative
```

#### Compute log likelihood

```python
def compute_log_likelihood(feature_matrix, sentiment, coefficient):
    indicator = (sentiment == +1)
    scores = np.dot(feature_matrix, coefficient)
    lp = np.sum((indicator-1)*scores - np.log(1. + np.exp(-scores)))
    return lp
```

#### Taking gradient step

```python
from math import sqrt
def logistic_regression(feature_matrix,sentiment,initial_coefficient,step_size,max_iter):
    coefficient = np.array(initial_coefficient)
    
    for itr in xrange(max_iter):
        predictions = predict_probability(feature_matrix,coefficient)
        indicator = (sentiment == +1)
        errors = indicator - predictions
        
        for j in xrange(len(coefficient)):
            derivative = feature_derivative(errors,feature_matrix[:,j])
            coefficient[j] = coefficient[j] + step_size*derivative
            
        if itr <= 15 or (itr <= 100 and itr % 10 == 0) or (itr <= 1000 and itr % 100 == 0) or (itr <= 10000 and itr % 1000 == 0) or itr % 10000 == 0:
            lp = compute_log_likelihood(feature_matrix,sentiment,coefficient)
            print 'iteration %*d: log likelihood of observed labels = %.8f' % (int(np.ceil(np.log10(max_iter))), itr, lp)
    return coefficient
```
